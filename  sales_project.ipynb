```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sales.csv')
df.head()
```

```python
# Check data types.
df.dtypes
```

```python
# Find total number of rows and columns.
df.shape
```

```python
# Find missing values.
df.isnull().sum().sum()
```

```python
# Remove duplicate rows.
df = df.drop_duplicates()
df
```

```python
# Select Sales column.
sales = df['Sales']
sales
```

```python
# Select Region and Profit.
df[['Region','Profit']]
```

```python
# Filter data where Sales > 3000.
filtered_sales = df[df['Sales'] >3000]
filtered_sales
```

```python
# Filter data where Region = North.
filtered_region = df[df['Region']=='North']
filtered_region
```

```python
# Calculate average Sales
cal_avgSales = df['Sales'].mean()
cal_avgSales
```

```python
# Find total Profit.
total_profit = df['Profit'].sum()
total_profit
```

```python
# Group by Region and calculate total Sales.
total_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
total_sales
```

```python
# Group by Category and find average Profit.
avg_profit = df.groupby('Category')['Profit'].mean()
avg_profit
```

```python
# Count how many orders per Payment_Mode.
df['Payment_Mode'].value_counts()
```

```python
# Find top 5 highest Sales records
sales_records = df.sort_values(by='Sales',ascending=False)
sales_records
```

```python
# Find least profitable orders.
least_profitable = df.sort_values(by='Profit')
least_profitable
```

```python
# Add new column: Profit_Per_Unit
df['Profit_Per_Unit'] = df['Profit'] / df['Quantity']
df
```

```python
# Sort dataset by Profit (descending)
sort_dataset = df.sort_values(by='Profit',ascending=False)
sort_dataset
```

## Matplotlib

```python
# Sales over Date.
sales_overdate = df.groupby('Date')['Sales'].sum()
sales_overdate

```

```python
sales_overdate.plot(kind='line',marker='o',color='g')
plt.title('Sales over Date')
plt.ylabel('Sales')
plt.grid()
plt.legend()
plt.show()
```

```python
# Sales by Region.
region_sales = df.groupby('Region')['Sales'].sum()
region_sales
```

```python
c=['r','g','y','b']
region_sales.plot(kind='bar',color=c)
plt.title('Sales by Region')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Profit by Category.
category_profit = df.groupby('Category')['Profit'].sum()
category_profit
```

```python
c = ['#6d83bb','#f49ac2','#978c59']
category_profit.plot(kind='bar',color=c)
plt.title('Profit by Category.')
plt.ylabel('Profit')
plt.show()
```

```python
# Quantity vs Sales.
quantity_sales = df.groupby('Quantity')['Sales'].sum()
quantity_sales
```

```python
quantity_sales = df[['Quantity','Sales']]
quantity_sales.plot(kind='scatter',x='Quantity',y='Sales')
plt.title('Quantity vs Sales')
plt.grid()
plt.show()
```

```python
# histogram of Customer Age
cust_age = df['Customer_Age']
cust_age
```

```python
cust_age.plot(kind='hist')
plt.title('Customer Age')
plt.xlabel('Age')
plt.grid()
plt.show()
```

```python
# pie chart of Payment_Mode.
payment = df['Payment_Mode'].value_counts()
payment
```

```python
payment.plot(kind='pie', autopct ='%1.2f%%')
plt.title("Payment Mode Distribution")
plt.ylabel("")
plt.show()
```

```python
# multiple lines (Sales & Profit).
multi_data = df.groupby('Date')[['Sales', 'Profit']].sum().reset_index()
multi_data
```

```python
multi_data.plot(x='Date',y=['Sales','Profit'],marker='o',linestyle='--')
plt.title("Sales & Profit Over Time")
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid()
plt.show()
```

```python
# Subplot (Sales & Profit)
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(df['Sales'],marker='o',color='c')
plt.title('Sales')
plt.grid()

plt.subplot(1,2,2)
plt.plot(df['Profit'],marker='*',color='g')
plt.title('Profit')
plt.grid()

plt.show()
```

```python
# Scatter plot (Sales vs Profit)
plt.scatter(df['Sales'],df['Profit'],color='#f49ac2')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
# Highlight highest sales point
max_sales = df['Sales'].max()
max_row = df[df['Sales'] == max_sales]
max_row
```

```python
plt.scatter(df['Sales'],df['Profit'])
plt.scatter(max_row['Sales'],max_row['Profit'])
plt.show()
```

```python
# Monthly Sales trend
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales
```

```python
monthly_sales.plot(kind='line',marker='^',color='r')
plt.title('Monthly Sales trend')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

## Seaborn

```python
# Sales by Region.
region_sales = df.groupby('Region')['Sales'].sum().reset_index()
region_sales
```

```python
sns.barplot(x='Region',y='Sales',data=region_sales,color='c',width=0.5)
plt.title('Sales by Region')

plt.show()
```

```python
# 	Payment_Mode
sns.countplot(x='Payment_Mode',data=df,width=0.5)
plt.title('Payment_Mode')
plt.show()
```

```python
# Histogram
sns.histplot(df['Profit'],kde=True)
sns.set_style('darkgrid')
plt.title('Profit')
plt.show()
```

```python
# Boxplot
sns.boxplot(x='Category',y='Sales',data=df)
plt.title('Sales by Category')
plt.show()
```

```python
# violin plot
sns.violinplot(x='Region',y='Profit',data=df,color='c')
plt.title('Profit by Region')
plt.show()
```

```python
# heatmap
corr = df.corr(numeric_only=True)

sns.heatmap(corr,annot=True)
plt.show()
```

```python
# pairplot
sns.pairplot(df.select_dtypes(include='number'))
plt.show()
```

```python
# barplot with hue
sns.barplot(x='Region',y='Sales',hue='Category',data=df)
plt.title('Sales by Region')
plt.show()
```
